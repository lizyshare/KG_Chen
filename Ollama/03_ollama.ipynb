```python
import requests
import json

def func(info):
    url = 'http://localhost:11434/api/chat'

    headers = {
        "Content-Type": "application/json"
    }

    data = {
        "model": "deepseek-r1:7b",
        "messages": [{"role": "user", "content": f"{info}。返回一个具有“歌名、歌手、作词、作曲、所属专辑、发行时间”内容的JSON对象，如果上文没有相关信息，对应的关键词置NULL"}],
        "stream": False,
        "format": {
            "type": "object",
            "properties": {
                "歌名": {
                    "type": "string"
                },
                "歌手": {
                    "type": "string"
                },
                "作词": {
                    "type": "string"
                },
                "作曲": {
                    "type": "string"
                },
                "专辑": {
                    "type": "string"
                },
                "发行时间": {
                    "type": "string"
                }
            },
            "required": [
                "歌名",
                "歌手",
                "作词",
                "作曲",
                "专辑",
                "发行时间"
            ]
        },
        "options": {
            "temperature": 0
        }
    }

    response = requests.post(url, headers=headers, data=json.dumps(data)).json()
    # print(type(response['message']['content']))
    if isinstance(response,dict):
        return json.loads(response['message']['content'])
    else:
        return None
```

```python
import pandas as pd
df = pd.read_csv("song_infos.csv")
names = df['歌曲名称'].to_list()
infos = df['歌曲信息'].to_list()

# 使用paddle进行关系提取
results = []

for i in range(0, len(names)):
    s = "["+str(i + 1) + "/"+str(len(names))+"]"

    name = names[i]
    info = infos[i]
    print(info)
    result = func(info)
    try:
        song_name = result['歌名']
        singer = result['歌手']
    except:
        print(s + name + ":获取关系失败")

    print(s + song_name)
    for r_name in ['歌手', '作词', '作曲', '专辑', '发行时间']:
        try:
            item = result[r_name]
            results.append([song_name, r_name, item])
        except:
            print(r_name + "关系提取失败")
    print("提取结束")   
```
