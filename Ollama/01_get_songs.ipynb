```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
# 获取陈奕迅的所有专辑
album_ids = []
album_names = []
# 不超过100个专辑，直接在一个页面显示
url_albums = "https://music.163.com/artist/album?id=2116&limit=100&offset=10"
headers = {
    'User-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'
}

res = requests.get(url=url_albums, headers=headers)
soup = BeautifulSoup(res.text, "lxml")

# 专辑id的特征为保存在a标签的href中
pattern = r'^/album\?id=\d+'
def is_album_a(tag):
    return tag.has_attr('href') and re.search(pattern=pattern, string=tag['href']) and len(tag['class']) != 1
for d in soup.find_all(is_album_a):
    album_ids.append(d['href'][10:])
    album_names.append(d.get_text())
```

```python
len(album_ids)
```

```python
# 获取各张专辑中的歌曲
all_songs = []
# 专辑页面
url_songs = "https://music.163.com/album?id="

def get_songs(album_id):
    res = requests.get(url=url_songs + album_id, headers=headers)
    soup = BeautifulSoup(res.text, "lxml")

    pattern = r'^/song\?id=\d+'
    def is_song_a(tag):
        return tag.has_attr('href') and re.search(pattern=pattern, string=tag['href'])
    for d in soup.find_all(is_song_a):
        all_songs.append(d.get_text())
    
for album_id in album_ids:
    get_songs(album_id=album_id)
```

```python
len(all_songs)
```

```python
# 对歌曲名进行过滤
cleaned_songs = []
for song in all_songs:
    if '(' in song:
        continue
    cleaned_songs.append(song)
```

```python
cleaned_songs
```

```python
# 保存到文件中
import pandas as pd
songs = pd.DataFrame(columns=["songs"], data=cleaned_songs)
songs.to_csv("all_songs.csv")
```
